```python
import json
import torch
from transformers import (AutoTokenizer, AutoModelForCausalLM,BitsAndBytesConfig,pipeline)
```

HF ACCOUNT CONFIGURATION


```python
config_data= json.load(open("config.json"))
HF_TOKEN = config_data["HF_TOKEN"]
```

```python
model_name="google/gemma-2-9b-it"
```

quantisation config

```python
bnb_config = BitsAndBytesConfig(load_in_4bit=True,
                                 bnb_4bit_use_double_quant=True,
                                 bnb_4bit_quant_type="nf4",
                                 bnb_4bit_compute_dtype=torch.bfloat16)
```

LOADING THE TOKENIZER AND THE MODEL

```python
tokenizer = AutoTokenizer.from_pretrained(model_name,token=HF_TOKEN)
tokenizer.pad_token=tokenizer.eos_token
```

```python
model = AutoModelForCausalLM.from_pretrained(model_name, quantization_config=bnb_config,device_map="auto",token=HF_TOKEN)
```

```python
text_generator = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    max_new_tokens=256,
)
```

```python
def get_response(prompt):
  sequences=text_generator(prompt)
  gen_text=sequences[0]['generated_text']
  return gen_text
```

```python
prompt="tell me about books?"
```

```python
gemma2_response=get_response(prompt)
```

```python
print(gemma2_response[len(prompt):])
```
